```python
import pandas as pd
```

```python
x=pd.read_csv(r"C:\Users\athar\.cache\kagglehub\datasets\aungpyaeap\supermarket-sales\versions\3\supermarket_sales.csv")
```

```python
df=pd.DataFrame(x)
```

```python
df.head(50)
```

```python
#Define the schema for the two dimension tables and one fact table.

# Customer Dimension
customer_dim = df[['Customer type', 'Gender']].drop_duplicates().reset_index(drop=True)
customer_dim['CustomerID'] = range(1, len(customer_dim) + 1)

# Product Dimension
product_dim = df[['Product line', 'Unit price']].drop_duplicates().reset_index(drop=True)
product_dim['ProductID'] = range(1, len(product_dim) + 1)

 
# Fact Table
fact_sales = df.merge(customer_dim, on=['Customer type', 'Gender']) \
                 .merge(product_dim, on=['Product line', 'Unit price']) \
                 [['Invoice ID', 'Date', 'Time', 'CustomerID', 'ProductID','Total', 'Tax 5%', 'Quantity']]
```

```python
import sqlite3

conn = sqlite3.connect('sales_data.db')

# Save tables to SQLite
customer_dim.to_sql('CustomerDimension', conn, if_exists='replace', index=False)
product_dim.to_sql('ProductDimension', conn, if_exists='replace', index=False)
fact_sales.to_sql('FactSales', conn, if_exists='replace', index=False)
print("Data loaded into SQLite successfully!")

#conn.close()
```

```python
for table in ['CustomerDimension', 'ProductDimension','FactSales']:
    print(f"Schema for {table}:")
    schema_query = f"PRAGMA table_info({table});"
    schema = pd.read_sql_query(schema_query, conn)
    print(schema)
```

```python
corrected_query = """
SELECT 
   "Product line","Unit price",ProductID  
 from
    ProductDimension pd  
 
"""

result = pd.read_sql_query(corrected_query, conn)
print(result)
```

```python
#Query: Total Sales by Product Line
a="""SELECT 
    pd."Product line", 
    SUM(fs.Total) AS Total_Sales, 
    SUM(fs.Quantity) AS Total_Quantity
FROM 
    FactSales fs
JOIN 
    ProductDimension pd
ON 
    fs.ProductID = pd.ProductID
GROUP BY 
    pd."Product line"
ORDER BY 
    Total_Sales DESC
    """
result = pd.read_sql_query(a, conn)
print(result)
```

```python
 #Query: Sales by Gender and Product Line
b="""
SELECT 
    cd.Gender, 
    pd."Product line", 
    SUM(fs.Total) AS Total_Sales
FROM 
    FactSales fs
JOIN 
    CustomerDimension cd
ON 
    fs.CustomerID = cd.CustomerID
JOIN 
    ProductDimension pd
ON 
    fs.ProductID = pd.ProductID
GROUP BY 
    cd.Gender, pd."Product line"
ORDER BY 
    cd.Gender, Total_Sales DESC;
"""

result = pd.read_sql_query(b, conn)
print(result)
```

```python
#3. Query: Top Customers by Total Spend
c="""SELECT 
    cd."Customer type", 
    cd.Gender, 
    fs.CustomerID, 
    SUM(fs.Total) AS Total_Spend,
    RANK() OVER (ORDER BY SUM(fs.Total) DESC) AS Rank
FROM 
    FactSales fs
JOIN 
    CustomerDimension cd
ON 
    fs.CustomerID = cd.CustomerID
GROUP BY 
    cd."Customer type", cd.Gender, fs.CustomerID
ORDER BY 
    Rank;
"""
result = pd.read_sql_query(c, conn)
print(result)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3
 
query = """
SELECT 
    pd."Product line", 
    SUM(fs.Total) AS Total_Sales, 
    SUM(fs.Quantity) AS Total_Quantity
FROM 
    FactSales fs
JOIN 
    ProductDimension pd
ON 
    fs.ProductID = pd.ProductID
GROUP BY 
    pd."Product line"
ORDER BY 
    Total_Sales DESC
"""

 
df = pd.read_sql_query(query, conn)

# Bar chart for Total Sales by Product Line
plt.bar(df['Product line'], df['Total_Sales'], color='skyblue', alpha=0.8, label='Total Sales')

# Line chart for Total Quantity by Product Line
plt.plot(df['Product line'], df['Total_Quantity'], color='orange', marker='o', label='Total Quantity')

 
plt.title('Total Sales and Quantity by Product Line', fontsize=14)
plt.xlabel('Product Line', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.legend()

# Display the graph
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3

 
result = pd.read_sql_query(b, conn)

 
pivot_data = result.pivot(index="Product line", columns="Gender", values="Total_Sales")

 
ax = pivot_data.plot(kind="bar", figsize=(10, 6), width=0.7, color=["skyblue", "orange"])

 
plt.title("Total Sales by Gender and Product Line", fontsize=16)
plt.xlabel("Product Line", fontsize=14)
plt.ylabel("Total Sales", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.legend(title="Gender", fontsize=12)
plt.tight_layout()

# Show the graph
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3

 
result = pd.read_sql_query(c, conn)

 
result = result.sort_values(by="Rank")

# Create a horizontal bar chart
plt.figure(figsize=(12, 8))

# Bar colors based on gender for better distinction
colors = result["Gender"].map({"Male": "skyblue", "Female": "lightcoral"})

plt.barh(
    result["CustomerID"].astype(str),
    result["Total_Spend"],
    color=colors,
    edgecolor="black",
)

# Add titles and labels
plt.title("Top Customers by Total Spend", fontsize=16)
plt.xlabel("Total Spend", fontsize=14)
plt.ylabel("Customer ID", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=10)
plt.legend(
    handles=[
        plt.Line2D([0], [0], color="skyblue", lw=6, label="Male"),
        plt.Line2D([0], [0], color="lightcoral", lw=6, label="Female"),
    ],
    title="Gender",
    fontsize=12,
)

plt.tight_layout()
plt.gca().invert_yaxis()  # Invert y-axis for ranking (Rank 1 at the top)
plt.show()
```
